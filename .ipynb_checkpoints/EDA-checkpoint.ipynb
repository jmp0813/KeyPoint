```python
import tensorflow as tf
import tensorflow_addons as tfa
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import albumentations as A

from PIL import Image
from tensorflow import keras

gpus = tf.config.experimental.list_physical_devices("GPU")
if gpus :
    try :
        tf.config.experimental.set_memory_growth(gpus[0], True)
    except RuntimeError as e :
        print(e)
```

```python
def plotKeyPoint(image, keypoint) :
    plt.figure(figsize=(120, 80))
    plt.subplot(5, 1, 3)
    for j in range(0,keypoint.shape[0], 2):
        plt.plot(keypoint[j], keypoint[j+1],'rx')
        plt.imshow(image)

def coordScaler(dataFrame, x_scale, y_scale,  x_columns, y_columns) :
    dataFrame.loc[:, dataFrame.columns[x_columns]] = ((dataFrame.loc[:, dataFrame.columns[x_columns]]) / 1920) * x_scale
    dataFrame.loc[:, dataFrame.columns[y_columns]] = ((dataFrame.loc[:, dataFrame.columns[y_columns]]) / 1080) * y_scale
    return dataFrame

trainTarget = pd.read_csv("./RawData/train_df.csv")
target_size = (256, 256)
batch_size = 32

x_coord = [i.endswith("x") for i in trainTarget.columns]
y_coord = [i.endswith("y") for i in trainTarget.columns]

trainTarget = coordScaler(trainTarget, target_size[0], target_size[1], x_coord, y_coord)
```

```python
data_paths = glob.glob('./RawData/train_imgs/*.jpg')
data_paths = [i.split("\\")[1] for i in data_paths]

train_paths = np.random.choice(data_paths, replace = False, size = int(len(data_paths) * 0.7))
valid_paths = np.setdiff1d(data_paths, train_paths, assume_unique = True)

test_paths = glob.glob('./RawData/test_imgs/*.jpg')
test_paths = [i.split("\\")[1] for i in test_paths]

def DataGenerator(train_paths):
    for i in train_paths.astype(str):
        img = keras.preprocessing.image.load_img("./RawData/train_imgs/" + i, target_size = target_size)
        
        img = keras.preprocessing.image.img_to_array(img)
        img /= 255.
        target = trainTarget.loc[trainTarget["image"] == i].iloc[:, 1:].values.reshape(-1)
        
        transformer = A.Compose([A.RandomBrightnessContrast(), 
                                 A.RandomRotate90(),
                                 A.HorizontalFlip(),
                                 A.VerticalFlip()],
                                 keypoint_params = A.KeypointParams(format='xy'))
        transformed = transformer(image = img, keypoints = target.reshape(-1, 2))
                
        yield (transformed["image"], np.array(transformed["keypoints"]).reshape(-1))
        
trainData = tf.data.Dataset.from_generator(DataGenerator, (tf.float32, tf.float32), (tf.TensorShape([target_size[0], target_size[1], 3]),tf.TensorShape([48])), args = [train_paths])
trainData = trainData.shuffle(1500).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)

validData = tf.data.Dataset.from_generator(DataGenerator, (tf.float32, tf.float32), (tf.TensorShape([target_size[0], target_size[1], 3]),tf.TensorShape([48])), args = [valid_paths])
validData = validData.shuffle(1500).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
```

```python
test = next(DataGenerator(train_paths))


plotKeyPoint(test[0], test[1])
```

```python
class SEBlock(keras.layers.Layer) :
    def __init__(self, input_shape, reduction_ratio = 16, **kwargs) :
        
        self.pooling = keras.layers.GlobalAveragePooling2D()
        self.dense_1 = keras.layers.Dense(input_shape // reduction_ratio, activation = "relu", use_bias = False)
        self.dense_2 = keras.layers.Dense(input_shape, activation = "sigmoid", use_bias = False)
        
        super(SEBlock, self).__init__(**kwargs)
    
    def call(self, x) :
        pool = self.pooling(x)
        dense_1 = self.dense_1(pool)
        output = self.dense_2(dense_1)
        output = tf.reshape(output, shape = (-1, 1, 1, x.shape[-1]))
        return output

class BottleNeckResidualUnit(keras.layers.Layer) :
    def __init__(self, filters, strides = 1, activation = "relu", **kwargs) :
        super().__init__(**kwargs)
        self.activation = keras.activations.get(activation)
        self.Layers = [
            keras.layers.Conv2D(filters // 4, 1, strides = strides, padding = "same", use_bias = False),
            keras.layers.BatchNormalization(),
            self.activation,
            keras.layers.Conv2D(filters // 4, 3, strides = strides, padding = "same", use_bias = False),
            keras.layers.BatchNormalization(),
            self.activation,
            keras.layers.Conv2D(filters, 1, strides = 1, padding = "same", use_bias = False),
            keras.layers.BatchNormalization()
        ]
        self.skip_layers = []
        
        if strides > 1 :
            self.skip_layers = [
                keras.layers.Conv2D(filters, 1, strides = strides, padding = "same", use_bias = False),
                keras.layers.BatchNormalization()
            ]
        
    def call(self, inputs) :
        x = inputs
        for layer in self.Layers :
            x = layer(x)
        skip_x = inputs
        for layer in self.skip_layers :
            skip_x = layer(skip_x)
        return self.activation(x + skip_x)
        
```

```python
InputLayer = keras.layers.Input(shape = (target_size[0], target_size[1], 3))
filters = 64

Conv = keras.layers.Conv2D(filters = filters, kernel_size = 3, strides = 2, padding = "same", activation = None)(InputLayer)
BN = keras.layers.BatchNormalization()(Conv)
Activ = keras.layers.LeakyReLU()(BN)
Drop = keras.layers.Dropout(0.5)(Activ)
Pool = keras.layers.MaxPooling2D()(Drop)
Resid = BottleNeckResidualUnit(filters)(Pool)
Block = SEBlock(input_shape = filters)(Resid)
Mult = keras.layers.Multiply()([Resid, Block])
Add = keras.layers.Add()([Pool, Mult])

Conv = keras.layers.Conv2D(filters = filters, kernel_size = 5, strides = 2, padding = "same", activation = None)(Add)
BN = keras.layers.BatchNormalization()(Conv)
Activ = keras.layers.LeakyReLU()(BN)
Drop = keras.layers.Dropout(0.5)(Activ)
Pool = keras.layers.MaxPooling2D()(Drop)
Block = SEBlock(input_shape = filters)(Pool)
Resid = BottleNeckResidualUnit(filters)(Pool)
Block = SEBlock(input_shape = filters)(Resid)
Mult = keras.layers.Multiply()([Resid, Block])
Add = keras.layers.Add()([Pool, Mult])

Conv = keras.layers.Conv2D(filters = filters, kernel_size = 7, strides = 2, padding = "same", activation = None)(Add)
BN = keras.layers.BatchNormalization()(Conv)
Activ = keras.layers.LeakyReLU()(BN)
Drop = keras.layers.Dropout(0.5)(Activ)
Pool = keras.layers.MaxPooling2D()(Drop)
Block = SEBlock(input_shape = filters)(Pool)
Resid = BottleNeckResidualUnit(filters)(Pool)
Block = SEBlock(input_shape = filters)(Resid)
Mult = keras.layers.Multiply()([Resid, Block])
Add = keras.layers.Add()([Pool, Mult])

Conv = keras.layers.Conv2D(filters = filters, kernel_size = 3, strides = 2, padding = "same", activation = None)(Add)
BN = keras.layers.BatchNormalization()(Conv)
Activ = keras.layers.LeakyReLU()(BN)
Drop = keras.layers.Dropout(0.5)(Activ)
Pool = keras.layers.MaxPooling2D()(Drop)
Block = SEBlock(input_shape = filters)(Pool)
Resid = BottleNeckResidualUnit(filters)(Pool)
Block = SEBlock(input_shape = filters)(Resid)
Mult = keras.layers.Multiply()([Resid, Block])
Add = keras.layers.Add()([Pool, Mult])

Flatten = keras.layers.Flatten()(Add)
OutputLayer = keras.layers.Dense(48, activation = None)(Flatten)

model.summary()

model = keras.models.Model(inputs = InputLayer, outputs = OutputLayer)
optimizer = keras.optimizers.RMSprop(learning_rate = 2.5e-4)
model.compile(optimizer = optimizer, loss = keras.losses.MeanSquaredError())
```

```python
keras.utils.plot_model(model)
```

```python
es = keras.callbacks.EarlyStopping(monitor = "val_loss", min_delta = 0.0005, patience = 30, restore_best_weights = True)
lr = keras.callbacks.ReduceLROnPlateau(monitor = "val_loss", factor = np.log10(3), patience = 5, verbose = 1, min_delta = 0.0005)

history = model.fit(trainData, epochs = 500, batch_size = batch_size, callbacks = [lr, es], validation_data = validData)
```

```python
def TestGenerator(paths):
    for i in paths.astype(str):
        img = keras.preprocessing.image.load_img("./RawData/test_imgs/" + i, target_size = target_size)
        
        img = keras.preprocessing.image.img_to_array(img)
        img /= 255.
        yield (img)

testData = tf.data.Dataset.from_generator(TestGenerator, tf.float32, tf.TensorShape([target_size[0], target_size[1], 3]), args = [test_paths])
testData = testData.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
```

```python
result = model.predict(testData)
plotData = list(testData)
plotResult = model.predict(tf.expand_dims(plotData[0][0], axis = 0))
plotKeyPoint(plotData[0][0], plotResult.reshape(-1))
```
